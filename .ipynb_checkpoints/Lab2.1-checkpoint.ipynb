```python
import polars as pl
```

```python
dh = pl.read_csv("data/DH.csv")
```

```python
slugging = pl.read_csv("data/slugging.csv")
```

```python
dh.head()
```

```python
slugging.head()
```

```python
long_for_plot = (dh.pivot(index='yearID', columns='lgID', values='ERA')
                 .with_columns((pl.col('AL') - pl.col('NL')).alias('AL - NL'))
                )

long_for_plot.head()
```

```python
long = long_for_plot.melt(
    id_vars="yearID",
    value_vars=["AL", "NL", "AL - NL"],
    variable_name="League",   # becomes: "AL", "NL", "AL - NL"
    value_name="ERA"        # numeric ERA values
)

long.head()
```

```python
from plotnine import (ggplot, aes, geom_line, geom_hline, geom_vline, annotate, theme, theme_classic, labs, scale_x_continuous)
import numpy as np

min_y, max_y = int(long["yearID"].min()), int(long["yearID"].max())
ten_year_ticks = range((min_y//10)*10, max_y+1, 10)

e = (ggplot(data = long)
 + geom_line(aes(x = "yearID",
                 y = "ERA",
                 color = "League"
                )
            )
 + theme_classic()
 + theme(figure_size=(10,4))
 + scale_x_continuous(breaks=ten_year_ticks, limits=(1950, 2025))
 + geom_vline(xintercept=1973, linetype="dashed", color="red")
 + geom_hline(yintercept=0, linetype="dotted")
 + annotate("text", x=1952, y=long["ERA"].min() + 1.25,
               label="DH introduced to the AL in 1973", ha="left", va="bottom", size=9)
 + labs(title="League ERA vs. yearID", x="yearID", y="League ERA")
)

e
```

```python

Slugging = slugging.filter(pl.col("lgID").is_in(["AL", "NL"]))
```

```python
wide = (
    Slugging.pivot(index="yearID", columns="lgID", values="Slugging", aggregate_function="first")
              .with_columns((pl.col("AL") - pl.col("NL")).alias("AL - NL"))
)
long = wide.melt(id_vars="yearID",
                 value_vars=["AL","NL","AL - NL"],
                 variable_name="League", value_name="Slugging%")
long = long.drop_nulls(subset=["Slugging%"])
long.head()
```

```python
from plotnine import (ggplot, aes, geom_line, geom_hline, geom_vline, annotate, theme, theme_classic, labs, scale_x_continuous, ggsave)
import numpy as np

s =(ggplot(data=long)
+ geom_line(aes(x = "yearID",
                y = "Slugging%",
                color = "League"
               )
           )
 + theme_classic()
 + theme(figure_size=(10,4))
 + scale_x_continuous(breaks=ten_year_ticks, limits=(1950, 2025))
 + geom_vline(xintercept=1973, linetype="dashed", color="red")
 + geom_hline(yintercept=0, linetype="dotted")
 + labs(title="League Slugging % vs. yearID", x="yearID", y="Slugging Percentage")
 + annotate("text", x=1952, y=long["Slugging%"].min() + 3.00,
            label="DH introduced to the AL in 1973", ha="left", va="bottom", size=9)
)

s
```

```python
ggsave(e, filename='ERAvsYear.png', dpi=300)
ggsave(s, filename='SluggingVsYear.png', dpi=300)
```
